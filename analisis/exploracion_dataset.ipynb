# Exploración del dataset crudo

Listado de componentes (señales) disponibles en `../data/raw.csv`.

```python
import pandas as pd

CSV_PATH = "../data/raw.csv"
df = pd.read_csv(CSV_PATH, nrows=0)

columnas = [c for c in df.columns if c != "Timestamp"]

print(f"Total componentes: {len(columnas)}")
print(f"Rango temporal disponible (leyendo primera y última línea)...\n")

for i, col in enumerate(columnas, 1):
    print(f"  {i:2d}. {col}")
```

```python
df_sample = pd.read_csv(CSV_PATH, parse_dates=["Timestamp"], nrows=5)
df_tail = pd.read_csv(CSV_PATH, parse_dates=["Timestamp"], skiprows=range(1, 3887241))

print(f"Primera muestra: {df_sample['Timestamp'].iloc[0]}")
print(f"Última muestra:  {df_tail['Timestamp'].iloc[-1]}")
print(f"Filas totales:   ~3.887.242")
print()
df_sample
```

```python
componentes = pd.DataFrame({
    "Componente": columnas,
    "Tipo": ["Potencia" if "Power" in c or "power" in c
             else "Tensión" if "Voltage" in c or "voltage" in c
             else "Frecuencia" if "Frequency" in c or "frequency" in c
             else "Temperatura" if "Temperature" in c or "temperature" in c
             else "Otro"
             for c in columnas],
})

print("Componentes por tipo:")
print(componentes.groupby("Tipo").size().sort_values(ascending=False).to_string())
print()
componentes
```
